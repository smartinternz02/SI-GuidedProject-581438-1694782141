```python
#import libraries and dataset
import pandas as pd
data = pd.read_csv('House Price India.csv')
data.head()
```

Uni-variate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram to get an idea about the underlying probability distribution of the variable.
plt.figure(figsize=(8, 6))
sns.histplot(data['living area'], kde=True)
plt.title('Distribution of Living Area')
plt.xlabel('Living Area')
plt.ylabel('Frequency')
plt.show()
```

```python
#using boxplot for understanding the spread and skewness of the data along with identifying outliers.
plt.figure(figsize=(8, 6))
sns.boxplot(x='number of bedrooms', y='Price', data=data)
plt.title('Number of Bedrooms vs. Price')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
# using countplot for visualizing the frequency of categorical variables.
plt.figure(figsize=(8, 6))
sns.countplot(x='condition of the house', data=data)
plt.title('Count of Houses by Condition')
plt.xlabel('Condition of the House')
plt.ylabel('Count')
plt.show()
```

Bi-Variate Analysis

```python
# using Scatter plot to understand the relationship between two continuous variables.
plt.figure(figsize=(8, 6))
sns.scatterplot(x='living area', y='Price', data=data)
plt.title('Living Area vs. Price')
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.show()
```

Multi-variate Analysis

```python
# Heatmap for correlation matrix
plt.figure(figsize=(12, 10))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pair plot for multivariate analysis
sns.pairplot(data, vars=['living area', 'number of bedrooms', 'number of bathrooms', 'Price'])
plt.suptitle('Pair Plot for Multivariate Analysis', y=1.02)
plt.show()
```
